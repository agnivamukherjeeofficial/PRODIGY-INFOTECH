### DATA PREPROCESSING

##### Handling null values

```python
data.isnull().sum()
```

```python
data.loc[data['Age'].isnull()]
```

```python
sns.histplot(data['Age'])
```

* As the histplot of Age column is slightly skewed, we are going to use median to replace the null values

```python
# calculates the median of the non null values in the age column and replace them in the place of the null values
data.loc[data['Age'].isnull(),'Age'] = np.median(data.Age.dropna())
```

```python
# since cabin column is a categorical feature, mode is used in order to replace the null values
data.loc[data['Cabin'].isnull(),'Cabin'] = stat.mode(data.Cabin.dropna())
```

```python
# Embarked column is also a categorical column, hence we are replacing the null values with mode
data.loc[data['Embarked'].isnull(),'Embarked'] = stat.mode(data.Embarked.dropna())
```

```python
# checking if all the null values has been handled
data.isnull().sum()
```

```python
# to check if there are any duplicate rows
data.duplicated().sum()
```

##### Encoding

```python
data['Sex'] = data.Sex.replace({'female':0,'male':1})
```

```python
data['Embarked'] = data.Embarked.replace({'S':0,'C':1,'Q':2})
```

```python
## For Name, tickets and Cabin column, label encoding has been used 
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
Name = encoder.fit_transform(data.Name)
data['Name'] = pd.DataFrame(Name)
```

```python
Ticket = encoder.fit_transform(data.Ticket)
data['Ticket'] = pd.DataFrame(Ticket)
```

```python
cabin = encoder.fit_transform(data.Cabin)
data['Cabin'] = pd.DataFrame(cabin)
```

```python
# final encoded data
data
```

##### Handling Outliers

```python
# plotting boxplot to check which columns has outliers

plt.figure(figsize=(20,20),facecolor='white')
plotnumber=1

for column in data:
        if plotnumber<=12:
            ax=plt.subplot(4,3,plotnumber)
            sns.boxplot(x=data[column],hue=data['Embarked'])
            plt.xlabel(column,fontsize=10)
            plt.ylabel('Embarked',fontsize=10)
        plotnumber+=1
plt.tight_layout()
```

* From the above boxplot we can see that there are outliers in Age and Fare column

```python
sns.distplot(data['Fare'])
```

* As from the above distplot we can see that the FARE column is skewed 

```python
# since the plot is skewed, IQR method is used for handling outliers in the age and fare column

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

IQR = Q3-Q1

min_limit = Q1-1.5*IQR
max_limit = Q3+1.5*IQR

outliers_count = ((data > max_limit) | (data < min_limit)).sum()

outliers_percentage = (outliers_count/len(data))*100

print('\n Sum of Outliers: \n',outliers_count)
print('\n Percentage of Outliers:\n',outliers_percentage)
```

```python
# handling outliers for AGE column

from scipy import stats

Q1 = np.percentile(data.Age,25)
Q3 = np.percentile(data.Age,75)

IQR = stats.iqr(data['Age'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['Age'] > max_limit) | (data['Age'] < min_limit),'Age'] = np.median(data['Age'])
```

```python
# handling outliers for FARE column

Q1 = np.percentile(data.Fare,25)
Q3 = np.percentile(data.Fare,75)

IQR = stats.iqr(data['Fare'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['Fare'] > max_limit) | (data['Fare'] < min_limit),'Fare'] = np.median(data['Fare'])
```

```python
Q1 = np.percentile(data.SibSp,25)
Q3 = np.percentile(data.SibSp,75)

IQR = stats.iqr(data['SibSp'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['SibSp'] > max_limit) | (data['SibSp'] < min_limit),'SibSp'] = np.median(data['SibSp'])
```
